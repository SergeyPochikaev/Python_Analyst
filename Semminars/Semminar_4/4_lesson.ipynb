### **3.2 Постройте график в относительных величинах**

```python
data = pd.crosstab(df.Company, df['Cpu_Company'])
data = data.drop('Samsung', axis=1).reset_index() # удаление и сброс индекса
data.head()
```

```python
plt.figure(figsize=(8, 6))

# определяем кол-во делений
n_ticks = np.arange(len(data['Intel']))

# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 0.4

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data['Intel'], width=w)
plt.bar(n_ticks + offset, data['AMD'], width=w)

plt.title('Взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров')
plt.xlabel('Производители ноутбуков')
plt.ylabel('Кол-во')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, data['Company'], rotation = 30);
```

## Задание 5*
### Постройте матрицу корреляций для таблицы**

```python
mat = df.corr()
mat
```

```python
mat = np.round(mat,2)
mat
```

    Корреляция происходит здесь по линейной зависимости:
- до 0,2 - очень слабая;
- до 0,5 - слабая;
- до 0,7 - средняя;
- до 0,9 - высокая;
- свыше 0,9 - очень высокая

```python
mat[np.abs(mat)<=0.2] = 0
mat
```

```python
sns.heatmap(mat,annot=True)
```

    Можно менять политру для тепловой карты

```python
sns.heatmap(mat,annot=True, cmap='viridis')
```

```python
upp_mat = np.triu(df.corr()) # отсечение верхней части матрицы
```

```python
sns.heatmap(mat,annot=True, cmap='viridis', mask= upp_mat);
```
